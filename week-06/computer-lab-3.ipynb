```python
Stress_F = np.loadtxt('stressFibers.csv',delimiter =',')
plt.figure()
plt.imshow(Stress_F)
plt.axis('off')
```

```python
#9.3
#covolution.py; part a
v = np.arange(-25, 26)
X, Y = np.meshgrid(v, v)
gauss_filter = np.exp(-.5*(X**2/2 + Y**2/45))

ax = Axes3D(plt.figure())
ax.plot_surface(X,Y,gauss_filter)
```

```python
# part b
laplace_filter = np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])
combined_filter = convolve(gauss_filter,laplace_filter,mode='valid')
ax = Axes3D(plt.figure())
ax.plot_surface(combined_filter,combined_filter,combined_filter)
```

```python
# Part c
Response_Stress = convolve(Stress_F, combined_filter)
plt.figure()
plt.imshow(Response_Stress)
plt.axis('off')


plt.figure()
plt.imshow(Response_Stress,vmin=0,vmax=.5*Response_Stress.max())
plt.axis('off')
```

The initial image produced is very difficult to see, as it seems to all be the same color. The second image is easier to see as it there is a contrast, thus allowing one to see the details of the image.
