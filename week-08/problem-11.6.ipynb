```python
m = 1
g = 3
zeta = 1
L = 1
#o = np.array([0,2,4,6])
o_1 = 0
o_2 = 2
o_3 = 4
o_4 = 6
pi =np.pi
#r = np.array([-pi,(pi/2), (3*pi)/2, (2*pi)])

theta_1 = np.linspace(-pi,pi,101)
theta_2 = np.linspace(-pi/2,pi/2,101)
theta_3 = np.linspace(-3*pi/2,3*pi/2,101)
theta_4 = np.linspace(-2*pi,2*pi,101)

o_rad1 = -((m*g*np.sin(theta_1)) + (zeta*L*o_1))/ (m*L)
o_rad2 = -((m*g*np.sin(theta_2)) + (zeta*L*o_2))/ (m*L)
o_rad3 = -((m*g*np.sin(theta_3)) + (zeta*L*o_3))/ (m*L)
o_rad4 = -((m*g*np.sin(theta_4)) + (zeta*L*o_4))/ (m*L)

Figure = plt.figure(figsize=(16,9))

#Nullclines (-pi,pi)
plt.plot(theta_1, o_rad1, color ='orange',lw=3, label= 'pi')
plt.axhline(y=0, color ='orange',lw=3,)
#Nullclines (-pi/2,pi/2)
plt.plot(theta_2, o_rad2, color ='red',lw=3, label= 'pi/2')

#Nullclines (-3pi/2,3pi/2)
plt.plot(theta_3, o_rad3, color ='green',lw=3, label= '3pi/2')

#Nullclines (-2pi,2pi)
plt.plot(theta_4, o_rad4, color ='blue',lw=3, label= '2pi')



#Meshgrid
x = np.linspace(-4*math.pi,4*math.pi,15)
y = np.linspace(-4*math.pi,4*math.pi,15)

X, Y = np.meshgrid(x,y)
vX = Y
vY = -X
plt.quiver(X,Y,vX,vY, pivot='mid')

#Center point
plt.plot(0,0,'og',ms=15)

#Bulls-Eye
plt.plot(-math.pi,0,'or',ms=20)

#labels
plt.xlabel('Angular Position,[rad]')
plt.ylabel('Angular Velocity, [rad/s]')

plt.legend(loc = 'upper right')
```

Within the plot, it is evident that different values provided for theta and omega shift the placement and length of each of the nullclines. This looks correct as the frequency and the value of theta are important factors for the bevahior of the lines.
