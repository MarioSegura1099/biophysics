## (a) Plot the phase portrait.

```python
m = 1
g = 3
xi = 1
L = 1
o = 0

theta = np.linspace(-np.pi,np.pi,101)

 
o_rad = -((m*g*np.sin(theta)) + (xi*L*o))/ (m*L)

Figure = plt.figure(figsize=(16,9))
#Nullclines
plt.plot(theta, o_rad, color ='orange',lw=3)
    #
plt.axhline(y=0, color ='orange',lw=3)

#Meshgrid
x = np.linspace(-math.pi,math.pi,15)
y = np.linspace(-math.pi,math.pi,15)

X, Y = np.meshgrid(x,y)
vX = Y
vY = -X
plt.quiver(X,Y,vX,vY, pivot='mid')

#Center point
plt.plot(0,0,'og',ms=15)

#Bulls-Eye
plt.plot(-math.pi,0,'or',ms=20)

#labels
plt.xlabel('Angular Position,[rad]')
plt.ylabel('Angular Velocity, [rad/s]')

```
