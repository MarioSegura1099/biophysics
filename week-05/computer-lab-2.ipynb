```python
#*************************************************************
# Mario Segura
# "Second Computer Lab", Chapter 7, Section 3

#Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Get a random number generator.
from numpy.random import default_rng
rng = default_rng()
rand = rng.random

# Factorial Function

import scipy.special as sc
from scipy.special import factorial
```

```python
P = 0.08
l = np.arange(0,15,1)
P_l = ((np.exp(-8) * 8**l) / sc.factorial(l))
print(P_l)
```

```python
N = 100
N_Flips = 100
M = np.zeros(N)

for i in range(N):
    Flip = 2*(rand(N_Flips) < 0.08) - 1
    H_Count = (Flip==1)
    Count = H_Count.sum()
    M[i] = Count
print(Count)
    
```

```python
N = 1000
N_Flips = 100
M = np.zeros(N)

for i in range(N):
    Flip = 2*(rand(N_Flips) < 0.08) - 1
    H_Count = (Flip==1)
    Count = H_Count.sum()
    M[i] = Count
print(Count)
```

```python
plt.hist(M)
#Title and Labels
plt.xlabel('Number of Bins')
plt.ylabel('Number of Heads in Trial')
plt.title("Distribution of Heads for N trials at 1,000")
```

```python
plt.hist(M)

N = 1000
l = np.arange(0,15,1)
P_l = ((np.exp(-8) * 8**l) / sc.factorial(l)) * N
plt.plot(P_l)
```

```python
N = 1000000
N_Flips = 100
M = np.zeros(N)

for i in range(N):
    Flip = 2*(rand(N_Flips) < 0.08) - 1
    H_Count = (Flip==1)
    Count = H_Count.sum()
    M[i] = Count

plt.hist(M)
#Title and Labels
plt.xlabel('Number of Bins')
plt.ylabel('Number of Heads in Trial')
plt.title("Distribution of Heads for N trials at 1,000,000")

l = np.arange(0,21,1)
P_l = (np.exp(-8) * 8**l / sc.factorial(l)) * N
plt.plot(P_l)
```

When comparing the two outcomes, it is evident that there is a similar relationship. Both graphs seem to follow a similar distribution, and if I were to increase the number of bins for the graph with the larger N, it would follow an even closer relationship. 

```python
N_Flips = 1000
Flips = (rand(N_Flips) < 0.08) - 1 + 1
One = np.nonzero(Flips)
Dif_1 = np.diff(One).flatten()

Counts = np.unique(Dif_1, return_counts=True)
```

```python
plt.bar(Counts[0], Counts[1])

#Title and Labels
plt.title('Frequency of Waiting Time')
plt.ylabel('Frequency')
plt.xlabel('Time')
```

The data found in the plot is rather difficult to find the characteristics of the distribution.

```python
plt.hist(Dif_1,bins=10)
#Title and Labels
plt.title('Frequency of Waiting Time')
plt.ylabel('Frequency')
plt.xlabel('Time')
```

The reduction of bins helped make the distibution easier to read and analyze. 

```python
plt.semilogy(Dif_1)
plt.title('Semi-log Plot of Distribution of Waiting Time')
```

```python
plt.loglog(Dif_1)
plt.title('Log-log Plot of Distribution of Waiting Time')
```

```python
Avg_Wait = Dif_1.mean()
Avg_Wait
```
