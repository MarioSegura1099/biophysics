```python
#*************************************************************
# Mario Segura
# "Problem 5.7"

#Libraries

import numpy as np
import matplotlib.pyplot as plt

# Get a random number generator.
from numpy.random import default_rng
rng = default_rng()
rand = rng.random
```

## (a) __Histogram of $1/x$__

```python
X = np.random.uniform(0,1,10000)
Inv_X = 1/X
count, bins, ignored = plt.hist(Inv_X,50, density = True, range = (0,25))
```

## (b) Log-log plot of frequencies

```python
plt.loglog(count)
```

## (c) Analysis and Conclusions

Based on the behavior of the histogram, it evident that there is a negative exponential behavior. When compared to the log-log plot, one can see that there is a negative slope that results from taking the log of an exponential. The behavior found in b makes sense with the behavior found in a.
