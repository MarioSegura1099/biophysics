```python
#*************************************************************
# Mario Segura
# "Problem 5.24"

#Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Data

df = pd.read_csv('djiweekly.csv',header=None)
df.columns = ['dji']

Week = np.arange(0,4224,1)
dji = df['dji'].to_numpy()
```

## (a) Time-series and histogram

```python
U_n = dji[1:]/dji[:-1]
y = np.log(U_n)
plt.figure(figsize = (9,6))
count,bins,ignore = plt.hist(y)
```

## (b) Gaussian approximation

```python
Dy = np.abs(bins[0]-bins[2])
Uy = 0

Sigma = 0.05
Ntot = len(y)

Ga = Ntot*(Dy/(Sigma*np.sqrt(2*3.14)))*np.exp(-(y-Uy)**2/(2*Sigma**2))
plt.plot(y,Ga,'or')
count,bins,ignore = plt.hist(y)
```

## (c) Comparison of logarithms

```python
log_c = np.log(count)
log_Ga = np.log(Ga)
plt.plot(log_c)

plt.title('Log Counts')
```

```python
plt.plot(log_Ga)
plt.title('Log Gaussian')
```

## (d) Power law approximation

```python
s = np.random.default_rng().standard_cauchy(5000)
s = s[(s > -10) & (s < 10)]
plt.hist(s, bins=100)
plt.show()
```

```python
log_s = np.log(s)
plt.plot(log_s)
plt.plot(log_Ga)
plt.title('Log Cauchy vs Log Gaussian')
```
