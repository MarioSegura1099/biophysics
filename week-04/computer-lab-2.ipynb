```python
#7.2--Part a

#Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Get a random number generator.
from numpy.random import default_rng
rng = default_rng()
rand = rng.random

#random walks

num_steps = 1000
num_trials = 1000
x_final = np.zeros(num_trials)
y_final = np.zeros(num_trials)
displacement = np.zeros(num_trials)

for i in range(num_trials):
    x_step = 2*(rand(num_steps) < 0.5) - 1
    y_step = 2*(rand(num_steps) < 0.5) - 1
    x = x_step.cumsum()
    y = y_step.cumsum()
    x_final[i] = x[-1]
    y_final[i] = y[-1]
    displacement[i] = np.sqrt(x[-1]**2 + y[-1]**2)

```

```python
# Part b
plt.hist(displacement)
plt.xlabel('Displacement')
plt.ylabel('Number of Occurances')
```

```python
# Part c

displacement_sq = displacement**2
plt.hist(displacement_sq)
plt.xlabel('Displacement')
plt.ylabel('Number of Occurances')
```

```python
# Part d
plt.figure(figsize = (12,7))
plt.subplot(2,1,1).semilogy(displacement,displacement_sq,"or")
plt.title('Semi-log Displacement vs Displacement**2')
plt.subplot(2,1,2).loglog(displacement,displacement_sq,"ob")
plt.title('Log-Log Displacement vs Displacement**2')
```

```python
#Part e
disp_sq_av = np.mean(displacement_sq)
print("The average value of displacment**2 is: {} ".format(disp_sq_av))
```

```python
#Part f

num_steps = 4000
num_trials = 1000
x_final = np.zeros(num_trials)
y_final = np.zeros(num_trials)
displacement = np.zeros(num_trials)

for i in range(num_trials):
    x_step = 2*(rand(num_steps) < 0.5) - 1
    y_step = 2*(rand(num_steps) < 0.5) - 1
    x = x_step.cumsum()
    y = y_step.cumsum()
    x_final[i] = x[-1]
    y_final[i] = y[-1]
    displacement[i] = np.sqrt(x[-1]**2 + y[-1]**2)
    
displacement_sq = displacement**2
disp_sq_av = np.mean(displacement_sq)
print("The average value of displacment**2 is: {} ".format(disp_sq_av))
```

```python
#How the mean-square displacement depends on the number of steps in a random walk

num_steps = np.array([2000,4000,6000,8000,10000])
num_trials = 1000
x_final = np.zeros(num_trials)
y_final = np.zeros(num_trials)
displacement = np.zeros(num_trials)
disp = np.array([np.zeros(num_trials),np.zeros(num_trials),np.zeros(num_trials),np.zeros(num_trials),np.zeros(num_trials)])
a = 0

for n in num_steps:
    for i in range(num_trials):
        x_step = 2*(rand(n) < 0.5) - 1 
        y_step = 2*(rand(n) < 0.5) - 1 
        x = x_step.cumsum() 
        y = y_step.cumsum()
        x_final[i] = x[-1] 
        y_final[i] = y[-1] 
        displacement[i] = np.sqrt(x[-1]**2 + y[-1]**2)
    disp[a] = displacement
    a = a+1

displacement_sq = disp**2
disp_sq_avg2000 = displacement_sq[0].mean()
disp_sq_avg4000 = displacement_sq[1].mean()
disp_sq_avg6000 = displacement_sq[2].mean()
disp_sq_avg8000 = displacement_sq[3].mean()
disp_sq_avg10000 = displacement_sq[4].mean()
disp_sq_avg = np.array([disp_sq_avg2000,disp_sq_avg4000,disp_sq_avg6000,disp_sq_avg8000,disp_sq_avg10000])

print("For num_steps = {} the average value of displacment^2 is {} ".format(num_steps[0],disp_sq_avg2000))
print("For num_steps = {} the average value of displacment^2 is {} ".format(num_steps[1],disp_sq_avg4000))
print("For num_steps = {} the average value of displacment^2 is {} ".format(num_steps[2],disp_sq_avg6000))
print("For num_steps = {} the average value of displacment^2 is {} ".format(num_steps[3],disp_sq_avg8000))
print("For num_steps = {} the average value of displacment^2 is {} ".format(num_steps[4],disp_sq_avg10000))
```

```python
#plots
plt.figure(figsize = (12,7))
plt.plot(num_steps,disp_sq_avg)
plt.title('# of Steps vs Mean-Square Displacement')
plt.ylabel('Mean-Square Displacement')
plt.xlabel('# of Steps')
```

```python
The plot correlates with the assumption of an increase in steps, leads to an increase in displacement. 
```
