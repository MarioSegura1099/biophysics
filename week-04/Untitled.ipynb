```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

from numpy.random import default_rng
rng = default_rng()
rand = rng.random
```

```python
Illness_inc = np.loadtxt("incidents.csv", delimiter=',')
```

```python
N = np.size(Illness_inc,0)
x_coord = Illness_inc[0:N:1,0]
y_coord = Illness_inc[0:N:1,1]
```

```python
plt.scatter(x_coord, y_coord, c="blue", marker='+')
plt.title('Plane of Infections')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
df = pd.read_csv("incidents.csv")
df.head()

fig, ax = plt.subplots()
df.plot(kind='scatter', x='x_coord', y='y_coord', ax=ax)

# Turn on the grid
ax.grid()
```
