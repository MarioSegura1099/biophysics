```python
#*************************************************************
# Mario Segura
# Problem 10.9

#Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Random

from numpy.random import default_rng
rng = default_rng()
rand = rng.random
```

```python
l0 = 0 
B_start = 1/37
B_stop = 1/6
B_birth = 5*B_stop
k_death = np.log(2)/50

N = 300
l = l0
a = 0 
T = 0
l_r = np.zeros(N)
t_r = np.zeros(N)
for n in range(N):
    rate = B_birth+ k_death*l
    t = - 1/rate * np.log(rand())
    
    if(T > 1/B_start):
        xi = B_birth/rate
        event = rand() < xi
        if event: 
             l = l+1
        else: 
            l = l-1
        a = a+1
        if(a > 1/B_stop):
            a = 0
            pass
    else:
        l = l
    T = T+t
    l_r[n] = l
    t_r[n] = T
    plt.plot(T,l,'.')
    plt.title('Death and Life')
```
