```python
#*************************************************************
# Mario Segura
# Problem 10.2

#Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#************************************
Data = np.load('colquhounData.npy')
Split = np.hsplit(Data,2)
Bins = Split[0].flatten()
Channels = Split[1].flatten()
```

## (a)

```python
N = Channels.sum()
N
Open_Channels = Channels.cumsum()
Open_Channels
```

## (b)

```python
#***********Number of Closed Channels******
plt.figure()
plt.title('Number of Closed Channels')
plt.plot(Bins,Channels)

#**********Number of Open Channels*********
plt.figure()
plt.title('Number of Open Channels')
plt.plot(Bins,Open_Channels)
```
