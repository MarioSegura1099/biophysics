## (a) Mechanical Oscillator

Create a phase portrait and streamlines for the mechanical oscillator.  (This is similar to Problem 13.1, but include nullclines this time.)  Discuss the qualitative behavior of the system.

```python
a = .05
b = .05
x = np.linspace(0, 10, 25)
y = np.linspace(0, 10, 25)

X, Y = np.meshgrid(x, y)
t = np.linspace(0, 10, 101)

y1 = 0 * t
y2 = 2*np.sin(1/(t**2))

dX = X
dY = -a * np.sin(X) - b * Y

plt.figure(dpi=200)
plt.quiver(X, Y, dY, dX, pivot='mid', scale=75, width=0.002, alpha=0.5)
plt.plot(t,y1,'g-', lw=1)
plt.plot(t,y2,'g-', lw=1)
plt.xlim (0,10)
plt.ylim(0,3)
```

The system seems to follow the behavior of a critically damped system. Since the system is going directly from one state to another, we see how there really is no oscillation over time; it goes from one state to the other with not oscillation. For example, the two buckets on page 382, once one gets full and the system then switches to the next bocket and slowly begins filling it.
