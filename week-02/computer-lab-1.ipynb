```python
# Array of time from 0 to 10

time= np.linspace(0,10,110)
```

```python
# declaration of variables
A= 70
B=0
alpha=9
beta=65

# Viral Load Model
viral_load = A * np.exp(-alpha * time) + B * np.exp(-beta*time)
```

```python
# plot of the two arrays

plt.plot(time, viral_load)

# graph details

plt.title('HIV Viral Load\n Concetration vs Time')
plt.xlabel('Time (days)')
plt.ylabel('Concentration')
```

```python
# Importing data

data = np.loadtxt("HIVseries.csv", delimiter=',')

# dividing data into two arrays

N = np.size(data,0)

Time = data[0:N:1,0]
concentration = data[0:N:1,1]
```

```python
# plotting data  

plt.scatter(Time,concentration, c = "green", marker='+', label='Concentration vs Days')

# graph details

plt.title('HIV Viral Load\n Concetration vs Time')
plt.xlabel('Time (days)')
plt.ylabel('Concentration')
plt.legend(loc='upper right')
```

```python
# combining data (Part A)

#**************plotting function******************** *************   
#creating similar time parameter

time= np.linspace(0,7,16)

# declaration of variables (Part B-fitting the curve)

A=87500   # A & B are the initial viral load
B=87500
alpha=.5  # the rate at which new cells are infected
beta=.5   # the rate at which virions  are removed from blood

# Viral Load Model

viral_load = A * np.exp(-alpha * time) + B * np.exp(-beta*time)


#Function
plt.plot(time, viral_load)

#************plotting data*******************************************************************

plt.scatter(Time,concentration, c = "green", marker='+', label='Concentration vs Days')

# graph details

plt.title('HIV Viral Load\n Concetration vs Time')
plt.xlabel('Time (days)')
plt.ylabel('Concentration')
plt.legend(loc='upper right')
```

```python
#**************************************Part C*********************

#**************plotting function******************** *************   
#creating similar time parameter

time= np.linspace(0,7,16)

# declaration of variables (Part B-fitting the curve)

A=87500   # A & B are the initial viral load
B=87500
alpha=.5  # the rate at which new cells are infected
beta=.5   # the rate at which virions  are removed from blood

# Viral Load Model*** 1/alpha

viral_load1 = A * np.exp(-(1/alpha) * time) + B * np.exp(-beta*time)

# viral load

viral_load = A * np.exp(-alpha * time) + B * np.exp(-beta*time)

#Function***1/alpha

plt.plot(time, viral_load1)

#Function
plt.plot(time, viral_load)

plt.legend(['-1/alpha','-alpha'])

# graph details

plt.title('HIV Viral Load\n Concetration vs Time')
plt.xlabel('Time (days)')
plt.ylabel('Concentration')
```

```python
While analyzing the two plots, it is evident that they have similar behaviour. However, the function with 1/alpha proves to have a slower infection rate. Since alpha represents the rate at which new cells are infected, by having 1/alpha we can see why this rate decreases. When looking at the latency period of HIV, and fitting this curve within those parameters, we find that this could not naturally exist within seven days, it would not be possible to have such a decrease in sevend days.  
```
